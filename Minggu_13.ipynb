```python
import matplotlib.pyplot as plt

import numpy as np
```

## Bar Chart

```python
df = pd.read_csv("./Data/data.csv")
df.plot(kind="bar",x="name",y="age",title="Age of Person",)
plt.show()
```

```python
df.plot(kind="bar",x="name",y=["num_children","num_pets"])
plt.show()
```

## Horizontal Bar Chart

```python
df.plot(kind="barh",x="name",y="age",title="Age of Person", color="red")
plt.show()
```

## Histogram

```python
df[["age"]].plot(kind="hist",bins=[0,20,40,60,80,100],rwidth=0.8)
plt.show()
```

## Pie Chart

```python
#Hitung distribusi frekuensi
#Plot ke pie chart

df2 = df["state"].value_counts()
df2.plot(kind = "pie")
plt.show()
```

## Stacked Bar Chart

```python
#Membuat tabel pivot

df3 = df.groupby(["gender","state"])["name"].size().unstack()
df3
```

```python
df3.plot(kind="bar",stacked=True)
plt.show()
```

```python
df.plot(kind = "bar", x = "name", y=["num_children","num_pets"], stacked = True)
plt.show()
```
